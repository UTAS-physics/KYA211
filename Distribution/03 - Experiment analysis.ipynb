## 1. Goal(s)

The purpose of this workshop is to determine if the period of a simple pendulum is independent of the starting angle. A recording of a simple pendulum will be used to extract positional coordinates of a pendulum bob as a function of time, and this in turn can be used to study the period of oscillation as a function of maximum height (or angle) of the pendulum.

## 2. Relevent background information

The simple pendulum is well described as a mass $m$ attached to a frictionless pivot point by a zero-mass string. The only force acting on the mass is that due to gravity (directed along $-\mathbf{y}$), and the tension force which acts in the direction between the mass and the pivot point. ### 2.1 Equations of motion

The system is most conveniently described in polar coordinates, which yields the equation of motion:

$$
\begin{bmatrix} \dot \theta \\ \ddot \theta \end{bmatrix} = \begin{bmatrix} \dot \theta \\ - \dfrac{g \sin \theta}{\ell}\end{bmatrix}
$$

which is a second-order, non-linear ordinary differential equation which is difficult to solve. It is commonly assumed that the angle $\theta$ is small, and under this condition, the equation is eminently solvable and yields a simple harmonic solution, that is, a constant period which is independent of the angle $\theta$.

## 3. What can we use for our experiment?

A simple pendulum was constructed using a brass mass, some nylon fishing line and a retort stand. The brass mass is spherical and so should act well as a point mass, the fishing line is sufficiently strong but extremely light as to not alter the centre of mass of the system, and the retort stand is simply there to provide a pivot point for the pendulum.

## 4. How are we going to do the experiment?

The experiment will be carried out by setting up a camera to record the motion of the pendulum, and the recording can in turn be analysed using software to extract the coordinates of the bob, and then we can extract the period from these data. Important considerations for the execution of the experiment:
- The pendulum should be released only with initial horizontal and vertical component (i.e. $x$ and $y$). From the nature of the recording, we cannot measure depth (the $z$ component) so if there is appreciable motion in this direction, we should discard the run and reset the motion
- Care must be taken to note the location of the pivot point, as the polar angle only makes sense if this is well known
- A high frame rate camera should be used as to produce sharp images and best deduce the position of the bob as a function of time

Initially, I had planned on making sure to provide a scale for both $x$ and $y$ in the video to ensure that I could convert the bob position in the video (measured in pixels) to the bob position in real space (meansured in metres); however, since we are only interested in the angular component of the motion, we need not worry about this. It is worth noting that we have now assumed that the scale factors for the positions in $x$ and $y$ are identical, which may not be true depending on the camera lens used, the position and angle of the camera, along with any image processing that is done either on the camera or in software, however, care was taken to ensure that everything was mutually rectilinear so these effects should be minimal, or at least obvious in the analysis if present.

## 5. The experiment

To verify if the period of oscillation is independent of angle, I want to ultimately produce a plot of period versus angle. One way to do this is to record many videos with different initial angles, calculate the period for each video and then plot these. The method does have a major flaw: if the period is not constant as a function of angle, as the pendulum slows due to drag, the period will change through the motion rendering the calculation inaccurate. I therefore decided to try and measure the "instantaneous" period versus angle, that is, have a single video and calculate the period for a given cycle, and see how this evolves as drag slows the pendulum and thus changes the initial angle. This is kind of like doing all the trails in an individual video.

Using a mobile phone with a 60 frames per second camera, I recorded the motion of a pendulum bob, and used the open source analysis software [tracker](https://opensourcephysics.github.io/tracker-website/) to get find the position of the centre of the bob as a function of time using the _autotracking_ feature, with the default settings used, but with care taken to select and edge feature (reflection of a window on the bob) - as recommended in the documentation - in order to get high-fidelity tracking. This data was exported to a `.csv` file which is to be analysed below.

The next task is to calculate the period. There are a few approaches to doing this, but the most straightforwad to me would be to use adjacent maxima and minima to calculate a "local" period. I could do this manually, but it seems like it should be an easy task to accomplish programatically.

I found a [StackOverflow](https://stackoverflow.com/questions/1713335/peak-finding-algorithm-for-python-scipy) thread which provided an example of using the `scipy` package, notably the `find_peaks` function from the `signal` library. The example code is shown below; I will need to adjust this to make it work for my data.

```python
from scipy.signal import find_peaks

x = np.sin(2*np.pi*(2**np.linspace(2,10,1000))*np.arange(1000)/48000) + np.random.normal(0, 1, 1000) * 0.15
peaks, _ = find_peaks(x, distance=20)
peaks2, _ = find_peaks(x, prominence=1)      # BEST!
peaks3, _ = find_peaks(x, width=20)
peaks4, _ = find_peaks(x, threshold=0.4)     # Required vertical distance to its direct neighbouring samples, pretty useless

plt.subplot(2, 2, 1)
plt.plot(peaks, x[peaks], "xr"); plt.plot(x); plt.legend(['distance'])
plt.subplot(2, 2, 2)
plt.plot(peaks2, x[peaks2], "ob"); plt.plot(x); plt.legend(['prominence'])
plt.subplot(2, 2, 3)
plt.plot(peaks3, x[peaks3], "vg"); plt.plot(x); plt.legend(['width'])
plt.subplot(2, 2, 4)
plt.plot(peaks4, x[peaks4], "xk"); plt.plot(x); plt.legend(['threshold'])

plt.show()
```
